```python
from sklearn.preprocessing import LabelEncoder
```

```python
lencode=LabelEncoder()
car['fuelsystem']=lencode.fit_transform(car[['fuelsystem']])
car['CarName']=lencode.fit_transform(car[['CarName']])
car['horsepower']=lencode.fit_transform(car[['horsepower']])
car['fueltype']=lencode.fit_transform(car[['fueltype']])
car['doornumber']=lencode.fit_transform(car[['doornumber']])
car['price']=lencode.fit_transform(car[['price']])
car['enginetype']=lencode.fit_transform(car[['enginetype']])
```

```python
from sklearn.preprocessing import MinMaxScaler
le=MinMaxScaler()
car['fuelsystem']=le.fit_transform(car[['fuelsystem']])
car['CarName']=le.fit_transform(car[['CarName']])
car['horsepower']=le.fit_transform(car[['horsepower']])

car['fuelsystem']=le.fit_transform(car[['fuelsystem']])
car['fueltype']=le.fit_transform(car[['fueltype']])
car['doornumber']=le.fit_transform(car[['doornumber']])
car['enginetype']=le.fit_transform(car[['enginetype']])
car['price']=le.fit_transform(car[['price']])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=car.iloc[:,0:6].values
y=car.iloc[:,6].values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
predict=model.predict(x_test)
```

```python
predict
```

```python
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
```

```python
mean_absolute_error(predict,y_test)
```

```python
mean_squared_error(predict,y_test)
```

```python
r2_score(predict,y_test)
```
